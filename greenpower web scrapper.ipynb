```python
url ='https://www.epa.gov/system/files/other-files/2021-07/toplists.xml'
response = get(url)
document = BeautifulSoup(response.text, 'html.parser')
# print(document)
```

```python
topLists = document.find_all('toplists')
```

```python
index = 1
columns = ["Partner Name", "kWh", "GP %","Industry","Green Power Resources"]
rows = []
for item in topLists:
    
    if(index > 100): 
        break
    
    partnerName = item.find('name').get_text()
    kwh = item.find('kwh').get_text()
    percentGP = item.find('percentgp').get_text()
    
    try:
        industry = item.find('orgtype').get_text()
    except:
        industry = ''
    
    gpresources = item.find('gpresources').get_text()
    
    rows.append([partnerName,kwh,percentGP,industry,gpresources])
    
    # print(index)
    # print('---------')
    # print(partnerName)
    # print(kwh)
    # print(percentGP)
    # print(industry)
    # print(gpresources)   
    # print('---------')
    index = index + 1
```

```python

partners = pd.DataFrame(rows,columns=columns)
partners.to_csv('partners.csv')
partners.head(100)
```
